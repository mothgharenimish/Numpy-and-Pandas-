```python
import pandas as pd
```

```python
students = pd.read_csv("StudentsPerformance.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns 
```

```python
# Step 1: Calculate total score per student
students["total_score"] = students[["math score", "reading score", "writing score"]].sum(axis=1)

# Step 2: Group by gender and sum all scores
gender_totals = students.groupby("gender")["total_score"].sum().reset_index()

# Step 3: Plot with Seaborn (fixing FutureWarning by using hue and legend=False)
plt.figure(figsize=(6,4))
sns.barplot(
    data=gender_totals,
    x="gender",
    y="total_score",
    hue="gender",          # Explicit hue
    dodge=False,           # Avoid double bars
    palette="pastel",
    legend=False           # Remove duplicate legend
)

# Add labels on bars
for i, v in enumerate(gender_totals["total_score"]):
    plt.text(i, v + 50, str(v), ha='center', va='bottom')

plt.xlabel("Gender")
plt.ylabel("Total Score (Math + Reading + Writing)")
plt.title("Total Scores by Gender")
plt.show()
```
